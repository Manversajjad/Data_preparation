## Problem Statement

Let us revisit our job fair problem. You notice that there are still many inconsistencies in the data even after removing missing values. Your task is to remove these inconsistencies and find the average salary of the students.

```python
#Importing the library
import numpy as np
import pandas as pd

#Loading the data
hiring_data = pd.read_csv('/content/hiring_data.csv')
```

Perform the previous steps

```python
#removing the rows that have more than four missing values
subset_df = hiring_data.dropna(thresh=5)
round(subset_df.salary.mean(),2)
```

Use this data to answer the following questions

**Q1:**  Previously, you calculated the average salary of the students who are hired. However, there are some duplicate rows in the data. What is the average salary of the students after filtering out the duplicate observations? (Rounded off to the nearest two decimal places)

a)   \$131,533.82

b)   \$127,890.46

c)   \$131,976.06

d)   \$137,071.15





```python
# Remove duplicates without using inplace, so you get a new DataFrame with duplicates removed
dedup = hiring_data.drop_duplicates()

# Calculate the average salary from the deduplicated DataFrame
avg_mean = dedup['salary'].mean()

# Round to two decimal places
avg_salary_rounded = round(avg_mean, 2)

avg_salary_rounded
```

**Q2:**  After removing the duplicate entries from the data, you start analyzing the column degree_b. You notice that this column contains some inconsistent values. There are supposed to be only three categories in the column degree_b: **‘Sci&Tech’, ‘Comm&Mgmt’, and ‘Others’.**

However, some entries under the categories Sci&Tech and Comm&Mgmt have been formatted as the abbreviations of these categories, **S&T** and **C&M**, respectively. This formatting error adds two more categories to the degree_b column. You decide to replace all instances of the abbreviations with the full forms of the corresponding categories such that the entire column contains only three different categories.

After this operation, what is the average salary of the students who had **Sci&Tech** degrees? (Rounded off to the nearest two decimal places)

   

a)   \$127,768.75

b)   \$135,014.29

c)   \$126,904.65

d)   \$139,017.65


```python
# write your code here
map_label = {'S&T': 'Sci&Tech', 'C&M': 'Comm&Mgmt'}
subset_df.replace(map_label,inplace = True)
print("The average salary of Sci&Tech students is:",round(hiring_data[hiring_data.degree_b == 'Sci&Tech'].salary.mean(),2))
```
